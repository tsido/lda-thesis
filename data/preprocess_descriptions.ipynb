```python
import pandas as pd
import csv
import re
import string

# Code for the preprocessing steps used to prepare the game descriptions to topic modelling
# libraries
```

```python
df = pd.read_csv('enriched_data.csv')

def clean_whitespace(row):
    word_array = row.split()
    trimmed_row = ' '.join(word_array)
    return trimmed_row

def strip_punctuation(row):
    translator = str.maketrans('', '', string.punctuation)
    trimmed_row = row.translate(translator)
    return trimmed_row

def lowercase(row):
    return row.lower()


# combine separate steps
def preprocess(row):
    trimmed_row = strip_punctuation(row)
    trimmed_row = clean_whitespace(trimmed_row)
    trimmed_row = lowercase(trimmed_row)
    return trimmed_row


df['TrimmedDescription'] = df['Description'].apply(preprocess)

df.to_csv('df_mallet.tsv', columns=['App_Name', 'TrimmedDescription'], sep="\t", header=False, quoting=csv.QUOTE_NONNUMERIC)

# This can be then run through mallet e.g. in Puhti:
# First, generate the mallet data file:

# $ bin/mallet import-file --input ../data/df_mallet.tsv --output ../data/apps.mallet  --remove-stopwords --keep-sequence

# Run the topic modelling: (we probably want to do more to capture the document topic relations
# as well)

#MALLET_MEMORY=6g bin/mallet train-topics --input ../data/apps.mallet --num-topics 20

```
